```python
import pandas as pd
import random
import numpy
```

```python
def generate_candidate_ranks(start, end, num):
    """
    start: beginning of rank range
    end: end of rank range
    num: number of ranks to generate per candidate
    
    random.randint: A random integer within the given range as parameters.
    """
    votes = [] 
    for j in range(num): 
        votes.append(random.randint(start, end)) 
    return votes 
```

```python
start_of_rank_range = 1
end_of_rank_range = 3
num_of_ranks = 1000

candidate_1 = generate_candidate_ranks(start_of_rank_range, end_of_rank_range, num_of_ranks)
candidate_2 = generate_candidate_ranks(start_of_rank_range, end_of_rank_range, num_of_ranks)
candidate_3 = generate_candidate_ranks(start_of_rank_range, end_of_rank_range, num_of_ranks)
```

```python
ballots = {'Candidate_1': candidate_1,
           'Candidate_2': candidate_2,
           'Candidate_3': candidate_3,
        }
```

```python
df = pd.DataFrame(ballots, columns = ['Candidate_0','Candidate_1', 'Candidate_2'])
```
