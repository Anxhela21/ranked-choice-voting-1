```python
import random
import pyrankvote
from pyrankvote import Candidate, Ballot
```

```python
random.seed(2)

kiss = Candidate('Bob Kiss')
montroll = Candidate('Andy Montroll')
wright = Candidate('Kurt Wright')
smith = Candidate('Dan Smith')
simpson = Candidate('James Simpson')
candidates = [kiss, montroll, wright, smith, simpson]
```

```python
weighted_choice = [5] * 32 + [2] * 21 + [3] * 20+ [1] * 17 + [4] * 10
#weighted_choice
```

```python
class Voter:
    def __init__(self, voter_id, candidates):
        self.voter_id = voter_id
        self.candidates = candidates
        self.vote=[]
        
    def num_choices(self):
        for i in range(100):
            return random.choice(weighted_choice)

    def cast_vote(self, num_choices, candidates):
        cand_list = list(range(len(candidates)))

        while num_choices>0:
            curr_choice=random.choice(cand_list)
            self.vote.append(self.candidates[curr_choice])
            cand_list.remove(curr_choice)
            num_choices = num_choices-1
```

```python

all_ballots=[]
ballots=[]
for v_id in range(1,100):#8980
    voter = Voter(v_id,candidates)
    choices=voter.num_choices()
    print("voter's id is:", voter.voter_id, "and he selects ", choices, "candidates")
    voter.cast_vote(choices, candidates)
    print("His vote is:", voter.vote, "\n\n")
    ballots.append(voter.vote)
    all_ballots.append(Ballot(voter.vote))
    
```

```python
election_result = pyrankvote.instant_runoff_voting(candidates, all_ballots)
```
