```python
import pandas as pd
import random
import numpy

#Random vote generator:
def Rand_vote(start, end, num): 
    votes = [] 
  
    for j in range(num): 
        votes.append(random.randint(start, end)) 

    return votes 


#Candidate 1
can1_num,can1_start,can1_end  = 1000000,1,5
can1 = Rand_vote(can1_start, can1_end, can1_num)

#Candidate 2
can2_num,can2_start,can2_end  = 1000000,1,5
can2 = Rand_vote(can2_start, can2_end, can2_num)

#Candidate 3
can3_num,can3_start,can3_end  = 1000000,1,5
can3 = Rand_vote(can3_start, can3_end, can3_num)

#Candidate 4
can4_num,can4_start,can4_end  = 1000000,1,5
can4 = Rand_vote(can4_start, can4_end, can4_num)

#Candidate 5
can5_num,can5_start,can5_end  = 1000000,1,5
can5 = Rand_vote(can5_start, can5_end, can5_num)
```

```python
ballots = {'Candidate_1': can1,
           'Candidate_2': can2,
           'Candidate_3': can3,
           'Candidate_4': can4,
           'Candidate_5': can5
        }

df = pd.DataFrame(ballots, columns = ['Candidate_1','Candidate_2', 'Candidate_3', 'Candidate_4', 'Candidate_5'])
```

```python
df.head()
```

```python
#it works ¯\_(ツ)_/¯
df = df[(df["Candidate_1"] != df['Candidate_2']) & (df["Candidate_1"] != df['Candidate_3']) & (df["Candidate_1"] != df['Candidate_4']) & (df["Candidate_1"] != df['Candidate_5'])]
df = df[(df["Candidate_2"] != df['Candidate_1']) & (df["Candidate_2"] != df['Candidate_3']) & (df["Candidate_2"] != df['Candidate_4']) & (df["Candidate_2"] != df['Candidate_5'])]
df = df[(df["Candidate_3"] != df['Candidate_2']) & (df["Candidate_3"] != df['Candidate_1']) & (df["Candidate_3"] != df['Candidate_4']) & (df["Candidate_3"] != df['Candidate_5'])]
df = df[(df["Candidate_4"] != df['Candidate_2']) & (df["Candidate_4"] != df['Candidate_3']) & (df["Candidate_4"] != df['Candidate_1']) & (df["Candidate_4"] != df['Candidate_5'])]
df = df[(df["Candidate_5"] != df['Candidate_2']) & (df["Candidate_5"] != df['Candidate_3']) & (df["Candidate_5"] != df['Candidate_4']) & (df["Candidate_5"] != df['Candidate_1'])]
```

```python
df = df.reset_index()
del df['index']
```

```python
wrong_votes = df.sample(frac =.1)

wrong_votes[['Candidate_1','Candidate_2','Candidate_3','Candidate_4','Candidate_5']] = 0
df.update(wrong_votes)

df.head(20)
```
